There are features that have constant values which can be dropped

Many of the featues are skewed, so it should be taken care of in scaling

The Outliers doesn't seem to be significant.

They do not appear to be explicitly causing the target to be in one group or another.

But removing outliers on every feature would likely lead to removal of all data

```python
# Clipping between 5 percentile and 95 percentile to avoid losing too much of data

lower_bound = df_features.quantile(0.05)
upper_bound = df_features.quantile(0.95)
```

```python
# Filter out outliers
outiers = (df_features < lower_bound) | (df_features > upper_bound)
```

```python
df_features[outiers] = np.nan
```

```python
# the columns which have constant values
df_features[df_features.columns[df_features.std()==0]].head()
```

```python
# dropping the columns with constant values
df_features.drop(columns=df_features.columns[df_features.std()==0], inplace=True)
```

```python
randomCols = np.random.randint(1,432,50)

plt.figure(figsize=(15, 35))
sns.set_theme(style='white')

for i, col in enumerate(df_features.columns[randomCols]):
    plt.subplot(20, 3, i+1)
    sns.histplot(x=df_features[col], kde=True)
    plt.xlabel(col, weight='bold')

plt.tight_layout()
```

```python
plt.figure(figsize=(15, 35))
sns.set_theme(style="darkgrid")

for i, col in enumerate(df_features.columns[randomCols]):
    plt.subplot(20, 3, i+1)
    sns.stripplot(x=df_features[col], y=df_target, hue=df_target, orient='h', palette=['violet', 'mediumblue'], linewidth=0.2, legend=False)
    plt.xlabel(col, weight='bold')

plt.tight_layout()
```
